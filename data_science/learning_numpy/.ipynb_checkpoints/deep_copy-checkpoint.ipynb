```python
import numpy as np
```

```python
countries = np.array(['India', 'Peru', 'USA', 'Nigeria', 'Pakistan', 'Germany'])
```

```python
loc = countries.copy()
loc
```

```python
loc is countries
```

```python
loc.base is countries
```

```python
loc[0] = 'Turkey'
loc
```

```python
countries
```

```python
loc.shape = 2, 3
```

```python
print('Contents of loc')
print(loc)
```

```python
print('Shape of Locations')
print(countries)
```

```python
# It makes an independent copy of the array.
# There is no link between the original and the copy.
```
