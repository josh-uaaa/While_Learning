```python
import numpy as np
```

```python
tech_companies = np.array([('IBM', 'Apple Inc.', 'Intel', 'Dell', 'Microsoft'),
                           ('New York', 'California', 'California', 'Texas', 'Washington')])
tech_companies
```

```python
tech_companies.shape
```

```python
# Similar to flatten(), except flatten() belongs to an array object and can only be used by Numpy arrays.
# The ravel() function belongs to the Numpy library and can be applied to any object that can be successfully parsed.
tech_companies.ravel()
```

```python
# Transpose.
tech_companies.T
```

```python
tech_companies.T.ravel()
```

```python
tech_companies.shape
```

```python
tech_companies.reshape(5, 2)
```

```python
np.arange(18).reshape(3, 6)
```
