```python
# Shallow Copy aka Views.
```

```python
import numpy as np
```

```python
countries = np.array(['India', 'Peru', 'USA', 'Nigeria', 'Pakistan', 'Germany'])
```

```python
loc_1 = countries.view()
loc_2 = countries.view()
```

```python
print(loc_1)
```

```python
print(loc_2)
```

```python
print('IDs for the arrays are different.\n')
print(f'ID for the countries is {id(countries)}\n')
print('IDs for the loc_views are...')
print(id(loc_1))
print(id(loc_2))
```

```python
loc_1 is countries
```

```python
loc_1.base is countries
```

```python
loc_2[3] = 'Ghana'
loc_2
```

```python
countries
```

```python
loc_1
```

```python
loc_1 = np.array(['France', 'Iran', 'Australia', 'Mexico', 'Cuba', 'Egypt'])
loc_1
```

```python
countries
```

```python
loc_2.shape = 2, 3
print(f'Shape of loc_2: {loc_2.shape}')
print(f'Contents:\n {loc_2}')
```

```python
print(f'Shape of countries array: {countries.shape}')
print(f'Contents:\n {countries}')
```

```python
# Views do not create a new copy of the data, instead referencing the original data, sharing the same memory location.
# Views are mutable, meaning you can modify the underlying data through the view (except reshaping).
# Views are useful for creating references to existing arrays without copying data, allowing for efficient manipulation and reshaping of arrays.
# Views are shallow copies.
```
