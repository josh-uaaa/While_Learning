```python
import numpy as np
```

```python
p = np.arange(12)
p
```

```python
print('Split the array to produce 4 equal-sized subarrays.')
np.split(p, 4)
```

```python
np. split(p, 5)
```

```python
print('Split the array at positions indicated 1-D array.')
np.split(p, [3, 8])
```

```python
array = np.array([('IBM', 'Apple Inc.', 'Intel', 'Sony', 'Dell', 'Akami'),
                  ('New York', 'California', 'California', 'Texas', 'Washington', 'Massachusetts')])
array
```

```python
# Horizontal Split (Columnwise).
x1, x2 = np.hsplit(array, 2)
```

```python
print(f'x1: {x1}\n')
print(f'x2: {x2}\n')
```

```python
np.hsplit(array, 6)
```

```python
np.hsplit(array, 4)
```

```python
# Vertical Split (Row-wise).
y1, y2 = np.vsplit(array, 2)
```

```python
print(f'y1: {y1}\n')
print(f'y2: {y2}\n')
```
