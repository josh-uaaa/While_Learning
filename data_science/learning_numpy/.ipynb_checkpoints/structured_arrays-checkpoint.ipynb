```python
import numpy as np
```

```python
emp_names = ['Danielle', 'Lorena', 'Manuel', 'Ryan', 'Teresa', 'Wes']
emp_ids = [1, 2, 3, 4, 5, 6]
emp_scores = [78.2, 57.50, 90, 77, 96.20, 87.37]
```

```python
# U16 = Unicode string of 16 characters.
# i4 = 4-byte int (int32).
# f8 = 8-byte float (float64).
emp_data = np.zeros(6, dtype={'names': ('Name', 'ID', 'Score'),
                              'formats': ('U16', 'i4', 'f8')})
```

```python
emp_data
```

```python
emp_data['Name'] = emp_names
emp_data['ID'] = emp_ids
emp_data['Score'] = emp_scores
emp_data
```
