```python
# Broadcasting - Allows operations on arrays with mismatched shapes.
# The smaller array is 'broadcasted' until it matches the size of the larger array.
# Broadcasting will not work with every single array, there are certain restrictions on the shapes of the arrays.
# The two arrays being operated on does not have to be of the same shape.
```

```python
# Broadcasting Scalars - Operations between scalars and arrays are always possible; the scalar is broadcasted over the array.
# Broadcasting Arrays - Operations between two arrays are possible only under certain conditions; smaller array broadcasted over the larger array.
```

```python
import numpy as np
```

```python
x = np.array([2, 4, 6, 8, 10])
y = np.array([5, 5, 5, 5, 5])
```

```python
x * y
```

```python
z = 10
```

```python
# Scalar Z broadcasted over Array X.
x * z
```

```python
ones = np.ones((3, 4))
```

```python
ones * z
```

```python
top_speeds = [184, 243, 192, 309, 257, 218]
weights = [1178, 1243, 1403, 1047, 1673, 1475]
```

```python
car_info = np.array([top_speeds, weights])
car_info
```

```python
# 1 km = 0.621371 miles.
# 1 kg = 2.20462 lbs.
conversion_factors = np.array([0.621371, 2.20462])
conversion_factors
```

```python
conversion_factors.shape
```

```python
car_info.shape
```

```python
car_info * conversion_factors
```

```python
new_factors = conversion_factors.reshape(2, 1)
new_factors
```

```python
new_factors.shape
```

```python
car_info * new_factors
```
