```python
import numpy as np
```

```python
p = np.array([9, 8, 7])
q = np.array([3, 2, 4])
```

```python
# Element-By-Element Operation.
# Assumption: Arrays are of the same size.
print(f"p + q = {p + q}")
print(f"p - q = {p - q}")
print(f"p * q = {p * q}")
print(f"p / q = {p / q}")
print(f"p % q = {p % q}")
```

```python
print(p % 2)
print(p > 8)
```

```python
x = np.array([[2, 1], [1, 3]])
y = np.array([[3, 2], [4, 2]])
print(f"x: \n{x}")
print(f"y: \n{y}")
```

```python
x + y
```

```python
x - y
```

```python
x * y
```

```python
# Dot Product.
x.dot(y)
```

```python
# Dot Product.
np.dot(x, y)
```

```python
print(x)
x *= 4
x
```

```python
# Aggregate Operations on 1D Array.
fleet_mileage = np.array([14130, 37234, 21892, 11479, 6890, 27981])
```

```python
fleet_mileage.sum()
```

```python
fleet_mileage.min()
```

```python
fleet_mileage.max()
```

```python
print(f"Mean: {fleet_mileage.mean()}")
```

```python
# Aggregate Operations on 2D Array.
num = np.arange(16).reshape(4, 4)
num
```

```python
num.sum()
```

```python
# Column Aggregate, Axis = 0.
num.sum(axis=0)
```

```python
# Row Aggregate, Axis = 1.
num.sum(axis=1)
```

```python
num.min(axis=1)
```

```python
num.mean(axis=1)
```
