```python
political_figures = ['Mustafa Kemal Ataturk', 'Franklin D. Roosevelt', 'Winston Churchill', 'Charles de Gaulle',
                     'Jawaharlal Nehru', 'Nelson Mandela', 'Deng Xiaoping']
```

```python
df_from_list = pd.DataFrame(political_figures)
df_from_list
```

```python
countries = ['Turkey', 'USA', 'UK', 'France', 'India', 'South Africa', 'China']
```

```python
politicians_tuples = list(zip(political_figures, countries))
politicians_tuples
```

```python
df_from_list_of_tuples = pd.DataFrame(politicians_tuples)
df_from_list_of_tuples
```

```python
df_from_list_of_tuples = pd.DataFrame(politicians_tuples, columns=['Politician Name', 'Country'])
df_from_list_of_tuples
```

```python
political_figures = ['Mustafa Kemal Ataturk', 'Franklin D. Roosevelt', 'Winston Churchill', 'Charles de Gaulle',
                     'Jawaharlal Nehru', 'Nelson Mandela', 'Deng Xiaoping']
countries = ['Turkey', 'USA', 'UK', 'France', 'India', 'South Africa', 'China']
```

```python
politicians_dict = {'Politician Name': pd.Series(political_figures),
                    'Country': pd.Series(countries)}
```

```python
politicians_df = pd.DataFrame(politicians_dict)
politicians_df
```

```python
politicians_df.to_csv('datasets/politicians.csv', index=False)
```
